```python
class Write2Dot:
    def __init__(
        self,
        d: "bool",
        start: "int",
        a: "list of list",
        ans: "list of list",
        f: "filename",
    ) -> "None":
        print("out graph", f)
        print("ZERO marks will be given if you don't write code here")
        with open(f, "w") as fl:
            fl.write("## Jagadeesh Vasudevamurthy ####\n")
            fl.write("graph g {\n")
            fl.write("\toverlap=false; splines=true\n")
            fl.write("\tedge [style=dotted, weight=10, len=.2]\n")
            colors = ["RED", "BLUE"]
            if len(ans) != 2:
                return
            for i, partition in enumerate(ans):
                fl.write("\t\tsubgraph cluster_" + colors[i] + " {\n")
                fl.write("\t\t" + colors[i] + " [pos=\"-1,0!\", color=" + colors[i].lower() + "]\n")
                for node in partition:
                    fl.write("\t\t" + str(node) + " -- " + colors[i] + "\n")
                fl.write("}\n")
            fl.write("edge [style=\"\", weight=1, len=1]\n")
            for group in a:
                for node in group:
                    if node < start:
                        continue
                    fl.write("\t" + str(start) + "--" + str(node) + "\n")
                start+=1
            fl.write("\tlabel= \"")
            for each in ans:
                fl.write(str(each))
            fl.write("\"\n}")
```
