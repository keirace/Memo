## Quick Sort O(nlogn)

```python
# use pivot, swap to get partition index
# sort left
# sort right

def quickSort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quickSort(arr, low, pi - 1)
        quickSort(arr, pi + 1, high)

def partition(arr, low, high):
    pivot = arr[high]
    j = low - 1
    for i in range(low, high):
        if arr[i] <= pivot :
            j += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[j+1], arr[high] = arr[high], arr[j+1]
    return j + 1

arr = [6,4,8,1,0]
quickSort(arr, 0, len(arr) - 1)
print(arr)
```

```python
import math

# merge sorted array
def _nlogn_time_constant_space(nums1: list, m: int, nums2:list, n: int):
    i = m
    for j in nums2:
        nums1[i] = j
        i += 1
    nums1.sort()
    k = m + n
    return nums1, math.ceil(k * math.log(k, 2))
print(_nlogn_time_constant_space([1,5,88, 0, 0], 3, [6, 9], 2))

def _n_time_n_space(nums1: list, m: int, nums2:list, n: int):
    temp = []
    work = 0
    i, j = 0, 0
    while i < m and j < n:
        work += 1
        if nums1[i] <= nums2[j]:
            temp.append(nums1[i])
            i += 1
        else:
            temp.append(nums2[j])
            j += 1
    while i < m:
        work += 1
        temp.append(nums1[i])
        i += 1
    while j < n:
        work += 1
        temp.append(nums2[j])
        j += 1
    for i in range(len(temp)):
        nums1[i] = temp[i]
    return nums1, work
print(_n_time_n_space([1,5,88, 0, 0], 3, [6, 9], 2))

def _n_time_constant_space(nums1: list, m: int, nums2:list, n: int):
    i, j, k = m - 1, n - 1, m+n - 1
    work = 0
    while i >= 0 and j >= 0:
        work += 1
        if nums1[i] < nums2[j]:
            nums1[k] = nums2[j]
            j -= 1
        else:
            nums1[k] = nums1[i]
            i -= 1
        k -= 1
    for t in range(j, -1, -1):
        work += 1
        nums1[k] = nums2[t]
        k -= 1
    return nums1, work
print(_n_time_constant_space([1,5,88, 0, 0], 3, [6, 9], 2))
```

## Fibonacci Time: O(2^n) Space: O(n)

```python
def fibo(i):
    if i < 2:
        return i
    return fibo(i - 1) + fibo(i - 2)

print(fibo(7))
```

```python
#dynamic programming
def fib (x: int):
    fib = [];
    for i in range(x+1):
        fib.append(0)
    fib[0] = 0;
    fib[1] = 1;
    for i in range(2, x+1, 1):
        fib[i] = fib[i - 1] + fib[i - 2];
    return fib[x]
		
print(fib(7))
```

```python
def reverse(a, ans):
    if a == 0:
        return ans
    ans = (a % 10) + (ans * 10)
    return reverse(a // 10, ans)
    
print(reverse(1986, 0))
# 1986 % 10
# 6 * 10, 198
# 68 * 10, 19
```

## Best time to buy and sell stocks

```python
def compute_profit(a: List[int], s: "int", b: "int") -> "int":
    n = len(a)
    if n == 0:
        return 0
    assert s >= 0 and s < n
    assert b >= 0 and b < n
    assert s >= b
    p = a[s] - a[b]
    return p

def _nlogn_time_logn_space(a: List[int]) -> "[sellday,buyday,work]":
    n = len(a)
    work = [0]
    if n == 0:
        return [0,0,0]
    return merge(a, 0, n - 1, work)

def merge(a, l, h, work):
    if l == h:
        return [l, h, work[0]]
    work[0] += 1
    mid = (h - l) // 2 + l
    left = merge(a, l, mid, work)
    right = merge(a, mid + 1, h, work)
    work[0] += 1
    min, max = -1, 0
    left_min = get_min_or_max_index(a, l, mid, min)
    right_max = get_min_or_max_index(a, mid + 1, h, max)
    cross_profit = compute_profit(a, right_max, left_min)
    left_profit = compute_profit(a, left[0], left[1])
    right_profit = compute_profit(a, right[0], right[1])

    max_profit = cross_profit
    max_profit_i = [right_max, left_min]
    if left_profit > cross_profit:
        max_profit = left_profit
        max_profit_i = [left[0], left[1]]
    if right_profit > cross_profit:
        max_profit = right_profit
        max_profit_i = [right[0], right[1]]
    return [max_profit_i[0], max_profit_i[1], work[0]]

def get_min_or_max_index(arr, l, h, minormax):
    val = arr[0]
    index = 0
    for i in range(l + 1, h + 1):
        if minormax == -1:
            if arr[i] < val:
                val = arr[i]
                index = i
        else:
            if arr[i] > val:
                val = arr[i]
                index = i
    return index

print("sellday, buyday, work")
arr = [5,10,4,6,12]
res = _nlogn_time_logn_space(arr)
print(res)
print("profit = ", compute_profit(arr, res[0], res[1]))
```

```python
def get_primes(n: int):
    k = n + 1
    primes = []
    primes.append(2)
    for i in range(3, k, 2):
        if not divisible_by_prime(i, primes):
            primes.append(i)
    return primes

def divisible_by_prime(i, primes):
    j = 0
    while primes[j] * primes[j] < i:
        if i % primes[j] == 0:
            return True
        j += 1
    return False

print(get_primes(10))
```
